# Keras

```python
from keras.models import Sequential
from keras.layers import Dense

from keras.optimizers import Adam
from keras.losses import categorical_crossentropy


import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
X,y = make_blobs(1000,centers=4, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
data = np.array([0, 4, 5, 0, 4, 4, 0, 0])
```

```python
ohe = OneHotEncoder(categories="auto")
```

```python
y_hot = ohe.fit_transform(y.reshape(-1,1)).todense()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y_hot, test_size=0.33, random_state=42)
```

```python


layer_1 = Dense(5, activation="relu", input_shape=(2,))
layer_2 = Dense(4, activation="softmax")
```

```python
model = Sequential()
model.add(layer_1)
model.add(layer_2)
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=10000, batch_size=200)
```

```python
model.evaluate(X_test, y_test)
```

```python
model.predict(X_test[:1])
```

```python
import pandas as pd
```

```python

data = pd.read_csv("mnist_train_small.csv")
```

```python
X = data.values[:,2:]
```

```python
y = data.values[:,1]
```

```python
ohe = OneHotEncoder(categories="auto")
```

```python
y_hot = ohe.fit_transform(y.reshape(-1,1)).todense()
y_hot[0].shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y_hot, test_size=0.33, random_state=42)
```

```python
layer_1 = Dense(28, activation="relu", input_shape=(784,))
layer_2 = Dense(21, activation="sigmoid")
layer_3 = Dense(14, activation="relu")
layer_4 = Dense(10, activation="softmax")
```

```python
model = Sequential()
model.add(layer_1)
model.add(layer_2)
model.add(layer_3)
model.add(layer_4)
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=200)
```

```python
model.evaluate(X_test, y_test)
```
