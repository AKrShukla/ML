```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def apply_filter(img, fil) :
    res = np.zeros([img.shape[0] - fil.shape[0] + 1, img.shape[1] - fil.shape[1] +1])
    for r in range(res.shape[0]) :
        for c in range(res.shape[1]) :
            cut = img[r:r+fil.shape[0], c:c+fil.shape[1]]
            res[r][c] = abs(np.sum(cut * fil))
    return res
    
```

```python
fil_h = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
fil_v = fil_h.T
fil_d = np.array([[1,1,0],[1,0,-1],[0,-1,-1]])
fil_blur = np.array([[1,1,1],[1,1,1],[1,1,1]])
fil_blur2 = np.ones((1,20))
```

```python
img = cv2.imread("nums/strange.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
small = cv2.resize(gray, (gray.shape[1]//2, gray.shape[0]//2))
plt.imshow(apply_filter(small,fil_blur2), cmap="gray")
```
