```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
bias = 100
X,y,coff = make_regression(100,n_features=1,bias=bias,coef=True,noise=10,random_state=42)
```

```python
plt.scatter(X,y)
```

```python
def hypothesis(bias,coff,X) :
    return X*coff + bias
```

```python
yh = hypothesis(bias,coff,X)
```

```python
plt.scatter(X,y)
plt.scatter(X,yh)
```
