```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
s = (178,316)
img = image.load_img("nums/strange.jpg", target_size=s)
```

```python
img
```

```python
np_img = image.img_to_array(img)
```

```python
plt.imshow(np_img.astype(int))
```

```python
generator = image.ImageDataGenerator(horizontal_flip=True,
                         rotation_range=70,
                         shear_range=0.2,
                         zoom_range=0.5,rescale=.1)
```

```python
batch = np.array([np_img])
```

```python
batch.shape
```

```python
gen = generator.flow(batch)
```

```python
while True :
    plt.figure()
    a=gen.next()
    plt.imshow(a[0].astype(int))
    plt.show()
    b=gen.next()
    break
    
```

```python
data = pd.read_csv("mnist_train_small.csv")
```

```python
X = data.values[:,2:].reshape(-1,28,28,1)
y = data.values[:,1]
```

```python
generator2 = image.ImageDataGenerator(horizontal_flip=True, 
                                      rotation_range=70, shear_range=0.2, zoom_range=0.5,rescale=.1)
```

```python
gen2 = generator2.flow(X,y)
```

```python
images, values = gen2.next()
```

```python
for img, val in zip(images,values) :
    plt.figure()
    plt.imshow(img.reshape(28,28).astype(int))
    plt.title(str(val))
    plt.show()
```
