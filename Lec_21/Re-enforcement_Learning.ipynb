```python
import gym
import numpy as np
```

```python
env = gym.make('CartPole-v1')
```

```python
inc = np.array([.01, 0, 0, 0])
dec = np.array([0.11, 0.1, 0.01, 0.1])
for i in range(1) :
    initial = env.reset()
    while(True) :
        env.render()
        state, reward, done, info = env.step(env.action_space.sample())
        env.unwrapped.state = initial
        if initial[0] < 1 :
            initial -= inc
        
        if initial[0] >=1.2 :
            initial-=dec
        if done :
            break
env.close()
```
