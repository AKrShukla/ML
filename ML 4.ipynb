# Sklearn

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(1000, centers=2, random_state=42)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1],s=4,label="Cluster 1",c=y)

# plt.xlabel("datarange")
# plt.ylabel("outcome")
plt.legend()
plt.title("Square and Cube")
# plt.xlim(-5,10)
# plt.ylim(-5,10)
plt.show()
```

```python
class knn :
    def __init__(self,k=10) :
        self.k=k
        self.X=None
        self.y=None
    def fit(self,X,y) :
        self.X=X
        self.y=y
    def predict_point(self,point) :
        distance=[]
        
        for x_item,y_item in zip(self.X,self.y) :
            dis = knn.distance(point,x_item)
            distance.append((dis,y_item))
        
        distance = sorted(distance)
        
        top = np.array(distance[:self.k])
        
        vals,counts = np.unique(top[:,1],return_counts=True)
        
        return (vals[np.argmax(counts)])
    
#     def predict(self,x) :
#         outs =[]
#         for point in x:
#             outs.append
    
    @classmethod
    def distance(cls, p1, p2) :
        return np.sum((p1-p2)**2)
            
            
```

```python
model = knn(4)
```

```python
model.fit(X,y)
```

```python
model.predict_point(np.array([-4,10]))
```
