```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
data1 = pd.read_csv("Diabetes_XTrain.csv")
```

```python
X_train=data1.values
```

```python
X_train
```

```python
data2 = pd.read_csv("Diabetes_YTrain.csv")
```

```python
y_train =  data2.values
y_train = y_train.reshape(-1,)

len(y_train)
```

```python
model = KNeighborsClassifier(17)
```

```python
model.fit(X_train, y_train)
```

```python
X_test = pd.read_csv("Diabetes_Xtest.csv").values
```

```python
X_test.dtype
```

```python
outcome = model.predict(X_test)
list(outcome)
```

```python
info = {
    "Outcome" : list(outcome)
}
```

```python
df = pd.DataFrame.from_dict(info)
```
