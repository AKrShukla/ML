```python
import pandas as pd
import numpy as np
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
data1 = pd.read_csv("Logistic_X_Train.csv")
```

```python
X_train=data1.values
```

```python
X_train
```

```python
data2 = pd.read_csv("Logistic_Y_Train.csv")
```

```python
y_train =  data2.values
y_train = y_train.reshape(-1,)

len(y_train)
```

```python
model = KNeighborsClassifier(10)
```

```python
model.fit(X_train, y_train)
```

```python
X_test = pd.read_csv("Logistic_X_Test.csv").values
```

```python
X_test.dtype
```

```python
outcome = model.predict(X_test)
list(outcome)
```

```python
chem = {
    "label" : list(outcome)
}
```

```python
df = pd.DataFrame.from_dict(chem)
```

```python
df
```

```python
df.to_csv("chem.csv")
```
