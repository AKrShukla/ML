```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url2 = "https://www.goodreads.com/quotes/tag/"
```

```python
tags_list = ['inspiration',
 'writing',
 'quotes',
 'religion',
 'success',
 'relationships',
 'life-lessons',
 'motivational',
 'knowledge',
 'time',
 'love-quotes',
 'spirituality',
 'science',
 'education',
 'books',
 'love',
 'life',
 'inspirational',
 'humor',
 'philosophy',
 'god',
 'inspirational-quotes',
 'truth',
 'wisdom',
 'romance',
 'poetry',
 'death',
 'happiness',
 'hope',
 'faith'
    ]
```

```python

# data = requests.get(url + tags_list[6])
len(tags_list)
```

```python
for tag in tags_list :
    i=1
    quote_dict = dict()
    q_t = list()
    q_a = list()
    url1 = url2+tag
    while True :
        url = url1+"?page="+str(i)
        data = requests.get(url)
        soup = BeautifulSoup(data.text,"lxml")
        quote_divs = soup.find_all("div", attrs={"class": "quote"})
        for quote_div in quote_divs :
            quoteText = quote_div.find("div", attrs={"class":"quoteText"})
            quote_info = quoteText.text
            quote_details = quote_info.strip().split("\n")
            quote_text = quote_details[0]
            quote_author = quote_details[3].strip()
            q_t.append(quote_text)
            q_a.append(quote_author)

    #     print(i)
        i=i+1
        if i>3 :
            break
    quote_dict = {"Quote" : q_t,
                 "Author" : q_a
                 }
    df = pd.DataFrame.from_dict(quote_dict)
    df.to_csv(tag+"-quotes.csv")
```

```python
for tag in tags_list :
    data = pd.read_csv("quotes/"+tag+"-quotes.csv")
    q_t = data["Quote"].values
    q_a = data["Author"].values
    q_tag = np.array([tag]*len(q_t))
    quote_dict = {"Quote" : q_t,
                 "Author" : q_a,
                  "Tag" : q_tag
                 }
    df = pd.DataFrame.from_dict(quote_dict)
    df.to_csv(tag+"-quotes.csv", index=False)
```

```python
frames = []
x = 0
for tag in tags_list :
    data = pd.read_csv("quotes/"+tag+"-quotes.csv")
    q_t = data["Quote"].values
    q_a = data["Author"].values
    q_tag = data["Tag"].values
    quote_dict = {"Quote" : q_t,
                 "Author" : q_a,
                  "Tag" : q_tag
                 }
    x1 = x
    x+=len(q_t)
    ind = [i for i in range(x1,x)]
    df = pd.DataFrame(quote_dict, index=ind)
    frames.append(df)
#     df.to_csv(tag+"-quotes.csv", index=False)
```

```python
result = pd.concat(frames)
```

```python
result.to_csv("quotes_data.csv", index=False)
```
