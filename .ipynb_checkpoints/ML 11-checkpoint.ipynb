# PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
X,y = make_blobs(1000,centers=10,n_features=10, random_state=42)
```

```python
# X.shape
# y.shape
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
model = PCA()
```

```python
model.fit(X)
```

```python
e_vec = model.components_
```

```python
X_dash=np.dot(X,e_vec.T)
```

```python
plt.scatter(X_dash[:,0],X_dash[:,1],c=y)
```

```python
# model.explained_variance_
```

```python
# model.explained_variance_ratio_
```

```python
plt.plot(model.explained_variance_ratio_)
```
